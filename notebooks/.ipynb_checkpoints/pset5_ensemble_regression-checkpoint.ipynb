````python
# Imports
import os
import urllib.request
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

# Spark imports
from pyspark.sql import SparkSession

# Scikit-learn
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV, TimeSeriesSplit
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Regresores base
from sklearn.linear_model import Ridge, LinearRegression
from sklearn.tree import DecisionTreeRegressor

# Ensemble methods
from sklearn.ensemble import (
    VotingRegressor, 
    BaggingRegressor,
    AdaBoostRegressor,
    GradientBoostingRegressor
)

# Boosting avanzado
import xgboost as xgb
import lightgbm as lgb
from catboost import CatBoostRegressor

# Interpretabilidad
import shap

plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
pd.set_option('display.max_columns', None)

# Seed para reproducibilidad
RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)

print("Imports completados.")
print(f"Random Seed: {RANDOM_SEED}")
````

````python
# Definición del problema

print("\n1. PROBLEMA Y OBJETIVO")
print("-" * 80)

print("""
Objetivo: Predecir total_amount en el momento del pickup

Evitar leakage:
- NO usar variables post-viaje (dropoff_datetime, trip_duration, avg_speed)
- Solo usar información disponible en el momento del pickup

Features permitidas:
- Temporales: pickup_datetime, pickup_hour, day_of_week, month, year
- Ubicación: pu_location_id, pu_zone, pu_borough
- Servicio: service_type, vendor_id, rate_code_id, payment_type
- Viaje: passenger_count, trip_distance
- Tarifas base: fare_amount, extra, mta_tax, improvement_surcharge
- Derivadas: is_rush_hour, is_weekend, base_fare_components

Target: total_amount (USD)

Partición temporal:
- Train: 2020
- Validación: 2021
- Test: 2022

Boosting Principal del Equipo: AdaBoost
""")
````

````python
# Configuración de Spark y PostgreSQL
print("\n2. CONFIGURACIÓN")
print("-" * 80)

# Variables de conexion a PostgreSQL
POSTGRES_HOST = os.getenv("POSTGRES_HOST")
POSTGRES_PORT = os.getenv("POSTGRES_PORT")
POSTGRES_DB = os.getenv("POSTGRES_DB")
POSTGRES_USER = os.getenv("POSTGRES_USER")
POSTGRES_PASSWORD = os.getenv("POSTGRES_PASSWORD")
PG_SCHEMA_ANALYTICS = os.getenv("PG_SCHEMA_ANALYTICS", "analytics")

# Verificar que existan
required_vars = {
    'POSTGRES_HOST': POSTGRES_HOST,
    'POSTGRES_PORT': POSTGRES_PORT,
    'POSTGRES_DB': POSTGRES_DB,
    'POSTGRES_USER': POSTGRES_USER,
    'POSTGRES_PASSWORD': POSTGRES_PASSWORD
}
missing_vars = [k for k, v in required_vars.items() if not v]
if missing_vars:
    raise ValueError(f"Faltan variables: {', '.join(missing_vars)}")
print("Variables de ambiente OK")

# Descargar driver JDBC de PostgreSQL
jar_dir = "/home/jovyan/jars"
os.makedirs(jar_dir, exist_ok=True)
postgres_jar_path = f"{jar_dir}/postgresql-42.7.1.jar"

if not os.path.exists(postgres_jar_path):
    print("Descargando PostgreSQL JDBC driver...")
    urllib.request.urlretrieve("https://jdbc.postgresql.org/download/postgresql-42.7.1.jar", postgres_jar_path)
    print(f"Driver descargado: {postgres_jar_path}")
else:
    print(f"Driver ya existe: {postgres_jar_path}")

# Crear SparkSession con driver PostgreSQL
spark = SparkSession.builder \
    .appName("NYC_Taxi_ML_Ensemble") \
    .config("spark.jars", postgres_jar_path) \
    .config("spark.sql.adaptive.enabled", "true") \
    .config("spark.sql.adaptive.coalescePartitions.enabled", "true") \
    .config("spark.driver.memory", "4g") \
    .config("spark.executor.memory", "4g") \
    .config("spark.sql.session.timeZone", "UTC") \
    .getOrCreate()
    
print(f"SparkSession creada")
print(f" - Version: {spark.version}")

# URL JDBC para PostgreSQL
JDBC_URL = f"jdbc:postgresql://{POSTGRES_HOST}:{POSTGRES_PORT}/{POSTGRES_DB}"

# Propiedades de conexión
pg_properties = {
    "user": POSTGRES_USER,
    "password": POSTGRES_PASSWORD,
    "driver": "org.postgresql.Driver"
}
print("Configuracion completada")
````

````python
# Carga de datos desde PostgreSQL
print("\n3. CARGA DE DATOS")
print("-" * 80)

# Query para PostgreSQL
base_query_template = f"""
(
    SELECT 
        total_amount,
        pickup_datetime,
        pickup_hour,
        pickup_dow as day_of_week,
        month, year,
        pu_location_id, pu_zone, pu_borough,
        service_type, vendor_id, rate_code_id, payment_type,
        passenger_count, trip_distance, store_and_fwd_flag,
        fare_amount, extra, mta_tax,
        improvement_surcharge, congestion_surcharge, airport_fee
    FROM {PG_SCHEMA_ANALYTICS}.obt_trips
    WHERE 
        total_amount IS NOT NULL AND total_amount > 0 AND total_amount < 500
        AND trip_distance IS NOT NULL AND trip_distance > 0 AND trip_distance < 100
        AND passenger_count IS NOT NULL AND passenger_count > 0 AND passenger_count <= 6
        AND pickup_datetime IS NOT NULL
        AND fare_amount IS NOT NULL AND fare_amount > 0
        AND year = {{year}}
    ORDER BY RANDOM()
    LIMIT 100000
) as subq
"""

# Cargar datos por año y combinar
dfs_spark = []
for year in [2020, 2021, 2022]:
    query = base_query_template.format(year=year)
    df_year = spark.read.jdbc(
        url=JDBC_URL,
        table=query,
        properties=pg_properties
    )
    dfs_spark.append(df_year)

# Combinar años
df_spark = dfs_spark[0].union(dfs_spark[1]).union(dfs_spark[2])
df_spark.cache()

# Convertir a pandas
df = df_spark.toPandas()

# Convertir pickup_datetime a datetime
df['pickup_datetime'] = pd.to_datetime(df['pickup_datetime'])

# Crear features derivadas si no existen
if 'is_weekend' not in df.columns:
    df['is_weekend'] = df['day_of_week'].isin([5, 6]).astype(int)

if 'is_rush_hour' not in df.columns:
    df['is_rush_hour'] = df['pickup_hour'].isin([7, 8, 9, 16, 17, 18, 19]).astype(int)

if 'base_fare_components' not in df.columns:
    df['base_fare_components'] = (df['fare_amount'] + 
                                   df['extra'] + 
                                   df['mta_tax'] + 
                                   df['improvement_surcharge'])

# ============================================================================
# PRINTS FINALES
# ============================================================================

print(f"\n{'='*80}")
print(f"DATOS CARGADOS EXITOSAMENTE")
print(f"{'='*80}")
print(f"Shape: {df.shape}")
print(f"Memoria: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")

print(f"\nColumnas disponibles:")
print(df.columns.tolist())

print(f"\nDistribución por año:")
year_dist = df['year'].value_counts().sort_index()
for year, count in year_dist.items():
    print(f"  {year}: {count:,} registros")

print(f"\nEstadísticas del target (total_amount):")
print(f"  Media: ${df['total_amount'].mean():.2f}")
print(f"  Mediana: ${df['total_amount'].median():.2f}")
print(f"  Std: ${df['total_amount'].std():.2f}")
print(f"  Min: ${df['total_amount'].min():.2f}")
print(f"  Max: ${df['total_amount'].max():.2f}")

# Verificar valores nulos
nulls = df.isnull().sum().sum()
if nulls > 0:
    print(f"\n Valores nulos encontrados: {nulls}")
    null_cols = df.isnull().sum()
    print(null_cols[null_cols > 0])
else:
    print(f"\n✓ Sin valores nulos")

# Mostrar primeras filas
print(f"\nPrimeras 3 filas:")
print(df.head(3))

print(f"\n{'='*80}")
print("✓ Carga de datos completada.")
print(f"{'='*80}")
````

````python
# Análisis exploratorio de datos

print("\n4. ANÁLISIS EXPLORATORIO")
print("-" * 80)

# Información básica
print("\nInformación del dataset:")
print(df.info())

print("\nEstadísticas descriptivas:")
print(df.describe())

# Visualización del target
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Histograma
ax1 = axes[0]
ax1.hist(df['total_amount'], bins=100, edgecolor='black', alpha=0.7, color='steelblue')
ax1.set_xlabel('Total Amount ($)')
ax1.set_ylabel('Frecuencia')
ax1.set_title('Distribución de total_amount')
ax1.axvline(df['total_amount'].mean(), color='red', linestyle='--', linewidth=2, label=f'Media: ${df["total_amount"].mean():.2f}')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Boxplot
ax2 = axes[1]
ax2.boxplot(df['total_amount'])
ax2.set_ylabel('Total Amount ($)')
ax2.set_title('Boxplot de total_amount')
ax2.grid(True, alpha=0.3)

# Total amount por año
ax3 = axes[2]
df.groupby('year')['total_amount'].mean().plot(kind='bar', ax=ax3, color='steelblue', edgecolor='black')
ax3.set_xlabel('Año')
ax3.set_ylabel('Total Amount Promedio ($)')
ax3.set_title('Total Amount Promedio por Año')
ax3.set_xticklabels(ax3.get_xticklabels(), rotation=0)
ax3.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()

print("\nDistribución de variables categóricas:")
categorical_cols = ['service_type', 'vendor_id', 'rate_code_id', 'payment_type', 'pu_borough']
for col in categorical_cols:
    if col in df.columns:
        print(f"\n{col}:")
        print(df[col].value_counts().head())
````

````python
# Definición de features y target

print("\n5. DEFINICIÓN DE FEATURES Y TARGET")
print("-" * 80)

# Features numéricas
numeric_features = [
    'passenger_count',
    'trip_distance',
    'fare_amount',
    'extra',
    'mta_tax',
    'improvement_surcharge',
    'pickup_hour',
    'day_of_week',
    'month',
    'is_weekend',
    'is_rush_hour',
    'base_fare_components'
]

# Agregar columnas adicionales si existen
optional_numeric = ['congestion_surcharge', 'airport_fee']
for col in optional_numeric:
    if col in df.columns:
        numeric_features.append(col)

# Features categóricas
categorical_features = [
    'service_type',
    'vendor_id',
    'rate_code_id',
    'payment_type',
    'pu_borough'
]

# Control de cardinalidad para pu_zone (Top-K + "Other")
if 'pu_zone' in df.columns:
    TOP_K_ZONES = 50
    top_zones = df['pu_zone'].value_counts().head(TOP_K_ZONES).index
    df['pu_zone_grouped'] = df['pu_zone'].apply(lambda x: x if x in top_zones else 'Other')
    categorical_features.append('pu_zone_grouped')
    print(f"✓ pu_zone agrupado: Top {TOP_K_ZONES} zonas + 'Other'")

# Target
target = 'total_amount'

print(f"\nFeatures numéricas ({len(numeric_features)}): ")
print(numeric_features)
print(f"\nFeatures categóricas ({len(categorical_features)}): ")
print(categorical_features)
print(f"\nTarget: {target}")

# Verificar que no hay leakage
prohibited_cols = ['dropoff_datetime', 'trip_duration_min', 'avg_speed_mph']
leakage_found = [col for col in prohibited_cols if col in df.columns]
if leakage_found:
    print(f"\n ADVERTENCIA: Posible leakage detectado: {leakage_found}")
else:
    print("\n✓ Sin leakage detectado")

# Seleccionar columnas relevantes
all_features = numeric_features + categorical_features
df_ml = df[all_features + [target, 'year']].copy()

print(f"\nDataset para ML:")
print(f" - Shape: {df_ml.shape}")
print(f" - Features totales: {len(all_features)}")
````

````python
# Split temporal por años

print("\n6. SPLIT TEMPORAL")
print("-" * 80)

# Train: 2020
df_train = df_ml[df_ml['year'] == 2020].copy()
X_train = df_train[all_features]
y_train = df_train[target]

# Validación: 2021
df_val = df_ml[df_ml['year'] == 2021].copy()
X_val = df_val[all_features]
y_val = df_val[target]

# Test: 2022
df_test = df_ml[df_ml['year'] == 2022].copy()
X_test = df_test[all_features]
y_test = df_test[target]

print(f"Train (2020):")
print(f" - Registros: {len(X_train):,}")
print(f" - Proporción: {len(X_train)/len(df_ml)*100:.1f}%")
print(f" - Target mean: ${y_train.mean():.2f}")

print(f"\nValidación (2021):")
print(f" - Registros: {len(X_val):,}")
print(f" - Proporción: {len(X_val)/len(df_ml)*100:.1f}%")
print(f" - Target mean: ${y_val.mean():.2f}")

print(f"\nTest (2022):")
print(f" - Registros: {len(X_test):,}")
print(f" - Proporción: {len(X_test)/len(df_ml)*100:.1f}%")
print(f" - Target mean: ${y_test.mean():.2f}")

# Visualización del split
fig, ax = plt.subplots(figsize=(10, 6))
split_data = pd.DataFrame({
    'Set': ['Train (2020)', 'Validation (2021)', 'Test (2022)'],
    'Registros': [len(X_train), len(X_val), len(X_test)],
    'Target_Mean': [y_train.mean(), y_val.mean(), y_test.mean()]
})

x_pos = np.arange(len(split_data))
bars = ax.bar(x_pos, split_data['Registros'], color=['#3498db', '#2ecc71', '#e74c3c'], 
              edgecolor='black', linewidth=1.5, alpha=0.8)
ax.set_ylabel('Número de Registros')
ax.set_title('Distribución del Split Temporal')
ax.set_xticks(x_pos)
ax.set_xticklabels(split_data['Set'])
ax.grid(True, alpha=0.3, axis='y')

for i, bar in enumerate(bars):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2., height,
            f'{int(height):,}\n(${split_data.iloc[i]["Target_Mean"]:.2f})',
            ha='center', va='bottom', fontweight='bold', fontsize=10)

plt.tight_layout()
plt.show()
````

````python
# Preprocesamiento común para todos los modelos

print("\n7. PREPROCESAMIENTO")
print("-" * 80)

from sklearn.impute import SimpleImputer

# Pipeline para features numéricas: imputar + escalar
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

# Pipeline para features categóricas: imputar + OHE
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])

print("Aplicando preprocesamiento...")
print(" - Imputación: Mediana para numéricas, 'missing' para categóricas")
print(" - Escalado: StandardScaler para numéricas")
print(" - Encoding: OneHotEncoder para categóricas")

X_train_processed = preprocessor.fit_transform(X_train)
X_val_processed = preprocessor.transform(X_val)
X_test_processed = preprocessor.transform(X_test)

# Verificar que no haya NaNs después del preprocesamiento
assert not np.isnan(X_train_processed).any(), " Hay NaNs en X_train después de preprocesamiento"
assert not np.isnan(X_val_processed).any(), " Hay NaNs en X_val después de preprocesamiento"
assert not np.isnan(X_test_processed).any(), " Hay NaNs en X_test después de preprocesamiento"

print("✓ Sin NaNs después del preprocesamiento")

# Obtener nombres de features después de OHE
feature_names = numeric_features.copy()
if hasattr(preprocessor.named_transformers_['cat'].named_steps['onehot'], 'get_feature_names_out'):
    cat_feature_names = preprocessor.named_transformers_['cat'].named_steps['onehot'].get_feature_names_out(categorical_features)
    feature_names.extend(cat_feature_names)

print(f"\nPreprocesamiento completado:")
print(f" - Train shape: {X_train_processed.shape}")
print(f" - Val shape: {X_val_processed.shape}")
print(f" - Test shape: {X_test_processed.shape}")
print(f" - Features totales después de OHE: {X_train_processed.shape[1]}")

print(f"\nEjemplos de features categóricas expandidas:")
if len(feature_names) > len(numeric_features):
    print(feature_names[len(numeric_features):len(numeric_features)+10])
````

````python
# Modelos baseline

print("\n8. BASELINE")
print("-" * 80)

# Baseline 1: Predecir la media
y_train_mean = y_train.mean()
y_val_pred_mean = np.full(len(y_val), y_train_mean)

rmse_baseline_mean = np.sqrt(mean_squared_error(y_val, y_val_pred_mean))
mae_baseline_mean = mean_absolute_error(y_val, y_val_pred_mean)
r2_baseline_mean = r2_score(y_val, y_val_pred_mean)

print(f"Baseline 1: Predecir la media (${y_train_mean:.2f})")
print(f" - RMSE: ${rmse_baseline_mean:.2f}")
print(f" - MAE: ${mae_baseline_mean:.2f}")
print(f" - R²: {r2_baseline_mean:.4f}")

# Baseline 2: Regresión lineal simple
print("\nBaseline 2: Regresión Lineal Simple")
lr_baseline = LinearRegression()
lr_baseline.fit(X_train_processed, y_train)
y_val_pred_lr = lr_baseline.predict(X_val_processed)

rmse_baseline_lr = np.sqrt(mean_squared_error(y_val, y_val_pred_lr))
mae_baseline_lr = mean_absolute_error(y_val, y_val_pred_lr)
r2_baseline_lr = r2_score(y_val, y_val_pred_lr)

print(f" - RMSE: ${rmse_baseline_lr:.2f}")
print(f" - MAE: ${mae_baseline_lr:.2f}")
print(f" - R²: {r2_baseline_lr:.4f}")

# Guardar mejores métricas de baseline
baseline_rmse = min(rmse_baseline_mean, rmse_baseline_lr)
baseline_mae = min(mae_baseline_mean, mae_baseline_lr)

print(f"\n✓ Mejor RMSE baseline: ${baseline_rmse:.2f}")
print(f"✓ Mejor MAE baseline: ${baseline_mae:.2f}")

# Guardar resultados
results_all = []
results_all.append({
    'Modelo': 'Baseline-Mean',
    'Tipo': 'Baseline',
    'RMSE_Val': rmse_baseline_mean,
    'MAE_Val': mae_baseline_mean,
    'R2_Val': r2_baseline_mean,
    'Tiempo_Train': 0
})
results_all.append({
    'Modelo': 'Baseline-LinearRegression',
    'Tipo': 'Baseline',
    'RMSE_Val': rmse_baseline_lr,
    'MAE_Val': mae_baseline_lr,
    'R2_Val': r2_baseline_lr,
    'Tiempo_Train': 0
})
````

````python
# Voting Regressor

print("\n9. VOTING REGRESSOR")
print("-" * 80)

# Definir modelos base para voting
ridge_base = Ridge(alpha=1.0, random_state=RANDOM_SEED)
tree_base = DecisionTreeRegressor(max_depth=10, random_state=RANDOM_SEED)
gb_base = GradientBoostingRegressor(n_estimators=50, max_depth=3, random_state=RANDOM_SEED)

voting_reg = VotingRegressor(
    estimators=[
        ('ridge', ridge_base),
        ('tree', tree_base),
        ('gb', gb_base)
    ]
)

print("\n✓ Modelos base seleccionados (3):")
print("  1. Ridge Regression (lineal, regularizado)")
print("  2. DecisionTree (no lineal, flexible)")
print("  3. GradientBoosting (ensemble, robusto)")
print("\n✓ Justificación:")
print("  - Diversidad: Combina modelos lineales, árboles y boosting")
print("  - Complementariedad: Ridge captura tendencias lineales,")
print("    Tree captura no linealidades, GB reduce bias")

# Pesos iguales (por defecto)
print("\n✓ Strategy: Promedio simple (voting='uniform')")
print("✓ Pesos: [1, 1, 1] (iguales para todos los modelos)")
print("-" * 80)


print("Entrenando Voting Regressor...")
print(" - Modelos base: Ridge, DecisionTree, GradientBoosting")

start_time = datetime.now()
voting_reg.fit(X_train_processed, y_train)
train_time = (datetime.now() - start_time).total_seconds()

y_val_pred_voting = voting_reg.predict(X_val_processed)

rmse_voting = np.sqrt(mean_squared_error(y_val, y_val_pred_voting))
mae_voting = mean_absolute_error(y_val, y_val_pred_voting)
r2_voting = r2_score(y_val, y_val_pred_voting)

print(f"\nResultados Voting Regressor:")
print(f" - RMSE: ${rmse_voting:.2f}")
print(f" - MAE: ${mae_voting:.2f}")
print(f" - R²: {r2_voting:.4f}")
print(f" - Tiempo entrenamiento: {train_time:.2f}s")

results_all.append({
    'Modelo': 'VotingRegressor',
    'Tipo': 'Ensemble-Voting',
    'RMSE_Val': rmse_voting,
    'MAE_Val': mae_voting,
    'R2_Val': r2_voting,
    'Tiempo_Train': train_time
})
````

````python
# Bagging vs Pasting

print("\n10. BAGGING VS PASTING")
print("-" * 80)

# Base learner: DecisionTreeRegressor
base_learner = DecisionTreeRegressor(max_depth=10, random_state=RANDOM_SEED)

# Bagging (bootstrap=True)
print("Entrenando Bagging...")
bagging_reg = BaggingRegressor(
    estimator=base_learner,
    n_estimators=100,
    max_samples=0.8,
    max_features=0.8,
    bootstrap=True,
    random_state=RANDOM_SEED,
    n_jobs=-1
)

start_time = datetime.now()
bagging_reg.fit(X_train_processed, y_train)
train_time_bagging = (datetime.now() - start_time).total_seconds()

y_val_pred_bagging = bagging_reg.predict(X_val_processed)

rmse_bagging = np.sqrt(mean_squared_error(y_val, y_val_pred_bagging))
mae_bagging = mean_absolute_error(y_val, y_val_pred_bagging)
r2_bagging = r2_score(y_val, y_val_pred_bagging)

print(f"Resultados Bagging:")
print(f" - RMSE: ${rmse_bagging:.2f}")
print(f" - MAE: ${mae_bagging:.2f}")
print(f" - R²: {r2_bagging:.4f}")
print(f" - Tiempo: {train_time_bagging:.2f}s")

# Pasting (bootstrap=False)
print("\nEntrenando Pasting...")
pasting_reg = BaggingRegressor(
    estimator=base_learner,
    n_estimators=100,
    max_samples=0.8,
    max_features=0.8,
    bootstrap=False,
    random_state=RANDOM_SEED,
    n_jobs=-1
)

start_time = datetime.now()
pasting_reg.fit(X_train_processed, y_train)
train_time_pasting = (datetime.now() - start_time).total_seconds()

y_val_pred_pasting = pasting_reg.predict(X_val_processed)

rmse_pasting = np.sqrt(mean_squared_error(y_val, y_val_pred_pasting))
mae_pasting = mean_absolute_error(y_val, y_val_pred_pasting)
r2_pasting = r2_score(y_val, y_val_pred_pasting)

print(f"Resultados Pasting:")
print(f" - RMSE: ${rmse_pasting:.2f}")
print(f" - MAE: ${mae_pasting:.2f}")
print(f" - R²: {r2_pasting:.4f}")
print(f" - Tiempo: {train_time_pasting:.2f}s")

# Comparación
print(f"\nComparación Bagging vs Pasting:")
print(f" - Diferencia RMSE: ${abs(rmse_bagging - rmse_pasting):.2f}")
print(f" - Mejor: {'Bagging' if rmse_bagging < rmse_pasting else 'Pasting'}")

results_all.append({
    'Modelo': 'Bagging',
    'Tipo': 'Ensemble-Bagging',
    'RMSE_Val': rmse_bagging,
    'MAE_Val': mae_bagging,
    'R2_Val': r2_bagging,
    'Tiempo_Train': train_time_bagging
})

results_all.append({
    'Modelo': 'Pasting',
    'Tipo': 'Ensemble-Pasting',
    'RMSE_Val': rmse_pasting,
    'MAE_Val': mae_pasting,
    'R2_Val': r2_pasting,
    'Tiempo_Train': train_time_pasting
})
````

````python
# AdaBoost con Grid Search (BOOSTING PRINCIPAL DEL EQUIPO)

print("\n11. ADABOOST (BOOSTING PRINCIPAL)")
print("=" * 80)

# Configurar TimeSeriesSplit
tscv = TimeSeriesSplit(n_splits=2)

# Grid de hiperparámetros para AdaBoost
ada_param_grid = {
    'n_estimators': [50, 100, 200],  
    'learning_rate': [0.05, 0.1, 0.5] 
}

print("Configurando Grid Search para AdaBoost...")
print(f" - Parámetros a buscar: {ada_param_grid}")
print(f" - CV: TimeSeriesSplit")
print(f" - Base learner: DecisionTree con max_depth=3")
print(f" - Combinaciones totales: {len(ada_param_grid['n_estimators']) * len(ada_param_grid['learning_rate']) * 2} fits")

ada_base = DecisionTreeRegressor(max_depth=3, random_state=RANDOM_SEED)
ada_reg = AdaBoostRegressor(estimator=ada_base, random_state=RANDOM_SEED)

ada_grid = GridSearchCV(
    ada_reg,
    ada_param_grid,
    cv=tscv,
    scoring='neg_root_mean_squared_error',
    n_jobs=-1,
    verbose=2  # Más detalle del progreso
)

print("\nEntrenando AdaBoost con Grid Search...")
print("(Estimado: 5-10 minutos...)")
start_time = datetime.now()
ada_grid.fit(X_train_processed, y_train)
train_time_ada = (datetime.now() - start_time).total_seconds()

print(f"\n{'='*80}")
print(f"✓ Mejores parámetros AdaBoost: {ada_grid.best_params_}")
print(f"✓ Mejor score CV: ${-ada_grid.best_score_:.2f}")
print(f"{'='*80}")

y_val_pred_ada = ada_grid.predict(X_val_processed)

rmse_ada = np.sqrt(mean_squared_error(y_val, y_val_pred_ada))
mae_ada = mean_absolute_error(y_val, y_val_pred_ada)
r2_ada = r2_score(y_val, y_val_pred_ada)

print(f"\nResultados AdaBoost en Validación:")
print(f" - RMSE: ${rmse_ada:.2f}")
print(f" - MAE: ${mae_ada:.2f}")
print(f" - R²: {r2_ada:.4f}")
print(f" - Tiempo: {train_time_ada:.2f}s ({train_time_ada/60:.1f} min)")

# Análisis de sensibilidad de hiperparámetros
print(f"\nAnálisis de sensibilidad (AdaBoost):")
cv_results = pd.DataFrame(ada_grid.cv_results_)
cv_results['mean_rmse'] = -cv_results['mean_test_score']
cv_results_sorted = cv_results.sort_values('mean_rmse')
print("\nTodas las combinaciones probadas:")
print(cv_results_sorted[['param_n_estimators', 'param_learning_rate', 'mean_rmse']].to_string(index=False))

results_all.append({
    'Modelo': 'AdaBoost',
    'Tipo': 'Boosting',
    'RMSE_Val': rmse_ada,
    'MAE_Val': mae_ada,
    'R2_Val': r2_ada,
    'Tiempo_Train': train_time_ada,
    'Best_Params': ada_grid.best_params_
})
````

````python
# Tabla comparativa de todos los modelos

print("\n16. COMPARATIVA COMPLETA")
print("-" * 80)

# Crear DataFrame con resultados
df_results = pd.DataFrame(results_all)

# Ordenar por RMSE en validación
df_results = df_results.sort_values('RMSE_Val').reset_index(drop=True)

print("\nTabla comparativa (ordenada por RMSE en Validación):")
print(df_results[['Modelo', 'Tipo', 'RMSE_Val', 'MAE_Val', 'R2_Val', 'Tiempo_Train']].to_string(index=False))

# Identificar el mejor modelo
best_idx = df_results['RMSE_Val'].idxmin()
best_model_name = df_results.loc[best_idx, 'Modelo']
best_model_rmse = df_results.loc[best_idx, 'RMSE_Val']

print(f"\n{'='*80}")
print(f" MEJOR MODELO: {best_model_name}")
print(f"   RMSE Validación: ${best_model_rmse:.2f}")
print(f"   MAE Validación: ${df_results.loc[best_idx, 'MAE_Val']:.2f}")
print(f"   R² Validación: {df_results.loc[best_idx, 'R2_Val']:.4f}")
print(f"{'='*80}")

# Visualización comparativa
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# RMSE por modelo
ax1 = axes[0, 0]
colors = ['red' if x == best_model_name else 'steelblue' for x in df_results['Modelo']]
bars = ax1.barh(df_results['Modelo'], df_results['RMSE_Val'], color=colors, edgecolor='black', alpha=0.7)
ax1.set_xlabel('RMSE ($)')
ax1.set_title('RMSE en Validación por Modelo')
ax1.axvline(baseline_rmse, color='orange', linestyle='--', linewidth=2, label=f'Baseline: ${baseline_rmse:.2f}')
ax1.legend()
ax1.grid(True, alpha=0.3, axis='x')

# MAE por modelo
ax2 = axes[0, 1]
bars = ax2.barh(df_results['Modelo'], df_results['MAE_Val'], color=colors, edgecolor='black', alpha=0.7)
ax2.set_xlabel('MAE ($)')
ax2.set_title('MAE en Validación por Modelo')
ax2.axvline(baseline_mae, color='orange', linestyle='--', linewidth=2, label=f'Baseline: ${baseline_mae:.2f}')
ax2.legend()
ax2.grid(True, alpha=0.3, axis='x')

# R² por modelo
ax3 = axes[1, 0]
bars = ax3.barh(df_results['Modelo'], df_results['R2_Val'], color=colors, edgecolor='black', alpha=0.7)
ax3.set_xlabel('R² Score')
ax3.set_title('R² en Validación por Modelo')
ax3.grid(True, alpha=0.3, axis='x')

# Tiempo de entrenamiento
ax4 = axes[1, 1]
tiempo_data = df_results[df_results['Tiempo_Train'] > 0]
bars = ax4.barh(tiempo_data['Modelo'], tiempo_data['Tiempo_Train']/60, color='seagreen', edgecolor='black', alpha=0.7)
ax4.set_xlabel('Tiempo (minutos)')
ax4.set_title('Tiempo de Entrenamiento')
ax4.grid(True, alpha=0.3, axis='x')

plt.tight_layout()
plt.show()

# Mejora sobre baseline
mejora_rmse = ((baseline_rmse - best_model_rmse) / baseline_rmse) * 100
mejora_mae = ((baseline_mae - df_results.loc[best_idx, 'MAE_Val']) / baseline_mae) * 100

print(f"\nMejora del mejor modelo sobre baseline:")
print(f" - Mejora RMSE: {mejora_rmse:.2f}%")
print(f" - Mejora MAE: {mejora_mae:.2f}%")
````

````python
# Evaluación del mejor modelo en Test

print("\n17. EVALUACIÓN EN TEST")
print("-" * 80)

print(f"Evaluando el mejor modelo ({best_model_name}) en el conjunto de Test...")

# Seleccionar el modelo correspondiente
if best_model_name == 'AdaBoost':
    final_model = ada_grid.best_estimator_
    X_test_final = X_test_processed
elif best_model_name == 'XGBoost':
    final_model = xgb_random.best_estimator_
    X_test_final = X_test_processed
elif best_model_name == 'LightGBM':
    final_model = lgb_random.best_estimator_
    X_test_final = X_test_processed
elif best_model_name == 'CatBoost':
    final_model = cat_random.best_estimator_
    X_test_final = X_test_processed
elif best_model_name == 'GradientBoosting':
    final_model = gb_grid.best_estimator_
    X_test_final = X_test_processed
elif best_model_name == 'Bagging':
    final_model = bagging_reg
    X_test_final = X_test_processed
elif best_model_name == 'Pasting':
    final_model = pasting_reg
    X_test_final = X_test_processed
elif best_model_name == 'VotingRegressor':
    final_model = voting_reg
    X_test_final = X_test_processed
else:
    final_model = lr_baseline
    X_test_final = X_test_processed

# Predicción en Test
y_test_pred = final_model.predict(X_test_final)

# Métricas en Test
rmse_test = np.sqrt(mean_squared_error(y_test, y_test_pred))
mae_test = mean_absolute_error(y_test, y_test_pred)
r2_test = r2_score(y_test, y_test_pred)

print(f"\n{'='*80}")
print(f"RESULTADOS FINALES EN TEST")
print(f"{'='*80}")
print(f"Modelo: {best_model_name}")
print(f" - RMSE Test: ${rmse_test:.2f}")
print(f" - MAE Test: ${mae_test:.2f}")
print(f" - R² Test: {r2_test:.4f}")
print(f"{'='*80}")

# Comparación Val vs Test
print(f"\nComparación Validación vs Test:")
print(f" - RMSE Val: ${best_model_rmse:.2f}")
print(f" - RMSE Test: ${rmse_test:.2f}")
print(f" - Diferencia: ${abs(rmse_test - best_model_rmse):.2f}")
diff_percent = abs(rmse_test - best_model_rmse) / best_model_rmse * 100
print(f" - Diferencia %: {diff_percent:.2f}%")
print(f" - Overfitting: {'Sí' if rmse_test > best_model_rmse * 1.1 else 'No'}")

# Visualización Val vs Test
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Scatter plot: Predicciones vs Real (Test)
ax1 = axes[0]
ax1.scatter(y_test, y_test_pred, alpha=0.3, s=10, color='steelblue')
ax1.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2, label='Línea perfecta')
ax1.set_xlabel('Total Amount Real ($)')
ax1.set_ylabel('Total Amount Predicho ($)')
ax1.set_title(f'Predicciones vs Real en Test\n{best_model_name}')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Residuales en Test
ax2 = axes[1]
residuals_test = y_test - y_test_pred
ax2.scatter(y_test_pred, residuals_test, alpha=0.3, s=10, color='coral')
ax2.axhline(y=0, color='r', linestyle='--', lw=2, label='Residual = 0')
ax2.set_xlabel('Total Amount Predicho ($)')
ax2.set_ylabel('Residuales ($)')
ax2.set_title(f'Residuales en Test\n{best_model_name}')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````

````python
# Análisis de error por buckets

print("\n18. ANÁLISIS DE ERROR POR BUCKETS")
print("-" * 80)

# Crear DataFrame con predicciones y real
df_test_analysis = X_test.copy()
df_test_analysis['total_amount_real'] = y_test.values
df_test_analysis['total_amount_pred'] = y_test_pred
df_test_analysis['error'] = df_test_analysis['total_amount_real'] - df_test_analysis['total_amount_pred']
df_test_analysis['abs_error'] = np.abs(df_test_analysis['error'])

# Análisis por rango de precio
print("\n1. Error por rango de precio:")
df_test_analysis['price_bucket'] = pd.cut(
    df_test_analysis['total_amount_real'],
    bins=[0, 10, 20, 30, 50, 100, np.inf],
    labels=['$0-10', '$10-20', '$20-30', '$30-50', '$50-100', '$100+']
)

bucket_analysis = df_test_analysis.groupby('price_bucket').agg({
    'abs_error': ['mean', 'median', 'std'],
    'total_amount_real': 'count'
}).round(2)
bucket_analysis.columns = ['MAE', 'Median_Error', 'Std_Error', 'Count']
print(bucket_analysis)

# Análisis por distancia
print("\n2. Error por rango de distancia:")
df_test_analysis['distance_bucket'] = pd.cut(
    df_test_analysis['trip_distance'],
    bins=[0, 2, 5, 10, 20, np.inf],
    labels=['0-2mi', '2-5mi', '5-10mi', '10-20mi', '20+mi']
)

distance_analysis = df_test_analysis.groupby('distance_bucket').agg({
    'abs_error': ['mean', 'median'],
    'total_amount_real': 'count'
}).round(2)
distance_analysis.columns = ['MAE', 'Median_Error', 'Count']
print(distance_analysis)

# Análisis por hora del día
print("\n3. Error por hora del día:")
hour_analysis = df_test_analysis.groupby('pickup_hour').agg({
    'abs_error': ['mean', 'median'],
    'total_amount_real': 'count'
}).round(2)
hour_analysis.columns = ['MAE', 'Median_Error', 'Count']
print(hour_analysis.head(12))

# Análisis por borough
print("\n4. Error por borough:")
if 'pu_borough' in df_test_analysis.columns:
    borough_analysis = df_test_analysis.groupby('pu_borough').agg({
        'abs_error': ['mean', 'median'],
        'total_amount_real': 'count'
    }).round(2)
    borough_analysis.columns = ['MAE', 'Median_Error', 'Count']
    print(borough_analysis)

# Visualización de análisis de buckets
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Error por rango de precio
ax1 = axes[0, 0]
bucket_analysis['MAE'].plot(kind='bar', ax=ax1, color='coral', edgecolor='black')
ax1.set_xlabel('Rango de Precio')
ax1.set_ylabel('MAE ($)')
ax1.set_title('Error Medio por Rango de Precio')
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=45, ha='right')
ax1.grid(True, alpha=0.3, axis='y')

# Error por distancia
ax2 = axes[0, 1]
distance_analysis['MAE'].plot(kind='bar', ax=ax2, color='lightblue', edgecolor='black')
ax2.set_xlabel('Rango de Distancia')
ax2.set_ylabel('MAE ($)')
ax2.set_title('Error Medio por Rango de Distancia')
ax2.set_xticklabels(ax2.get_xticklabels(), rotation=45, ha='right')
ax2.grid(True, alpha=0.3, axis='y')

# Error por hora
ax3 = axes[1, 0]
hour_analysis['MAE'].plot(kind='line', ax=ax3, marker='o', color='green', linewidth=2)
ax3.set_xlabel('Hora del Día')
ax3.set_ylabel('MAE ($)')
ax3.set_title('Error Medio por Hora del Día')
ax3.grid(True, alpha=0.3)

# Distribución de errores absolutos
ax4 = axes[1, 1]
ax4.hist(df_test_analysis['abs_error'], bins=50, edgecolor='black', alpha=0.7, color='purple')
ax4.set_xlabel('Error Absoluto ($)')
ax4.set_ylabel('Frecuencia')
ax4.set_title('Distribución de Errores Absolutos')
ax4.axvline(df_test_analysis['abs_error'].mean(), color='red', linestyle='--', 
            linewidth=2, label=f'Media: ${df_test_analysis["abs_error"].mean():.2f}')
ax4.legend()
ax4.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()
````
